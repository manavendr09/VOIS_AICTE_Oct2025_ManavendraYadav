```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/Dataset/1730285881-Airbnb_Open_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.duplicated().value_counts()
```

```python
df[df['neighbourhood group']=='brookln']
```

```python
df.drop_duplicates(inplace=True)
df.drop(['house_rules','license'], axis=1, inplace=True, errors='ignore')
df.rename(columns={'price':'price_$','service fee':'service_fee_$'},inplace=True)
df.dropna(inplace=True)
df['price_$']=df['price_$'].astype(float)
df['service_fee_$']=df['service_fee_$'].astype(float)
df['id']=df['id'].astype(str)
df['host id']=df['host id'].astype(str)
df['last review']=pd.to_datetime(df['last review'])
df['Construction year'] = df['Construction year'].fillna(0).astype(int)
df.loc[df['neighbourhood group'] =='brookln','neighbourhood group']='Brooklyn'
df=df.drop(df[df['availability 365']>500].index)
```

```python
df.duplicated().value_counts()
```

```python
df.info()
```

```python
df.describe()
```

```python
property_types =df['room type'].value_counts().to_frame()
property_types
```

```python

room_type_bar=plt.bar(property_types.index,property_types.loc[:,"count"]);
plt.bar_label(room_type_bar,labels=property_types.loc[:,"count"],padding=4);
plt.ylim([0,90000])
plt.ylabel('Room Type Count')
plt.xlabel('Room Type')
plt.title('property types and thier count in data set');
```

```python
hood_group=df['neighbourhood'].value_counts().to_frame()
hood_group
```

```python
top_10_hoods = hood_group.head(10)

plt.figure(figsize=(10, 6))
hood_group_bar = plt.bar(top_10_hoods.index, top_10_hoods.loc[:, "count"])
plt.bar_label(hood_group_bar, labels=top_10_hoods.loc[:, "count"], padding=4)
plt.ylabel('Neighbourhood Count')
plt.xlabel('Neighbourhood')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Neighbourhoods by Count');
plt.tight_layout()
plt.show()
```

```python
avg_price = df.groupby('neighbourhood')['price_$'].mean().to_frame()
top_10_avg_price = avg_price.loc[top_10_hoods.index]

plt.figure(figsize=(10, 6))
avg_price_bar = plt.bar(top_10_avg_price.index, top_10_avg_price.loc[:, "price_$"])
plt.bar_label(avg_price_bar, labels=round(top_10_avg_price.loc[:,"price_$"], 2), padding=4);
plt.ylim([00, 1000])
plt.ylabel('Average Price')
plt.xlabel('Neighbourhood')
plt.xticks(rotation=45, ha='right')
plt.title('Average Price in Top 10 Neighbourhoods');
plt.tight_layout()
plt.show()
```

```python
df.groupby(df['Construction year'])['price_$'].mean().to_frame().plot();
plt.xlabel('Construction Year')
plt.ylabel('Average Price')
plt.title('Average Price by Construction Year');
```

```python
top_10_hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending=False).nlargest(10).to_frame()
top_10_hosts_bar = plt.bar(top_10_hosts.index,top_10_hosts['calculated host listings count']);
plt.bar_label(top_10_hosts_bar,labels=top_10_hosts['calculated host listings count'],label_type='edge',padding=4)
plt.ylim([0,150000])
plt.ylabel('Total Listings Count')
plt.xlabel('Host name')
plt.xticks(rotation=90, ha='right')
plt.title('Top 10 Hosts by Total Listings Count');
plt.tight_layout()
plt.show()
```

```python
review=df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review
```

```python
review_bar=plt.bar(review.index,review['review rate number']);
plt.bar_label(review_bar,labels=review['review rate number'],label_type='edge',padding=4);
plt.ylim([0,4])
plt.ylabel('Average Review Rate')
plt.xlabel('Host Identity Verified')
plt.title('Average Review Rate by Host Identity Verified');
#
```

```python
base_color=sns.color_palette()[0]
sns.boxenplot(data=df,x="host_identity_verified",y="review rate number",color=base_color);
plt.xlabel('Host Verification Status')
plt.ylabel('Price')
plt.title('Price Distribution by Host Verification Status');
```
